```python
import json
from glob import glob
from os.path import join as pjoin
import cv2
```

```python
def extract_children_elements(element, all_elements, ele_id):
    element['id'] = ele_id
    all_elements.append(element)
    if 'children' in element:
        element['children_id'] = []
        for child in element['children']:
            ele_id += 1
            element['children_id'].append(ele_id)
            ele_id = extract_children_elements(child, all_elements, ele_id)    
        del element['children']
    return ele_id
```

```python
def visualize_elements(img, elements):
    board = img.copy()
    for ele in elements:
        print(ele)
        bounds = ele['bounds']
        cv2.rectangle(board, (bounds[0], bounds[1]), (bounds[2], bounds[3]), (0,255,0), 3)
        cv2.imshow('ele', cv2.resize(board, (board.shape[1] // 3, board.shape[0] // 3)))
        cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
data_dir = 'E:\\Mulong\\Datasets\\gui\\rico\\combined\\all'
```

```python
jfiles = glob(pjoin(data_dir, '*.json'))
```

```python
for jfile in jfiles:
    name = jfile.split('\\')[-1].split('.')[0]
    imgfile = pjoin(data_dir, name + '.jpg')
    print(imgfile)
    
    jdata = json.load(open(jfile, 'r'))
    img = cv2.resize(cv2.imread(imgfile), (1440,2560))
    
    ele_root = jdata['activity']['root']
    all_eles = []
    extract_children_elements(ele_root, all_eles, 0)    

    break
```
